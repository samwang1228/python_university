```python
import winsound as wd
import pygame as pg
import time
Frequency = {1:261, 2:293, 3:329, 4:349, 5:392, 6: 440, 7:493 \
             ,8:523,9:587,10:659,11:698,12:784,13:880,14:988 \
            ,15:1046,16:1175,17:1318,18:1397,19:1568,20:1760,21:1976}
music = [1,1,5,5,6,6,5,4,4,3,3,2,2,1] #小星星

music2=[1,1,1,3,5,5,5,5,6,6,6,8,5,4,4,4,6\
        ,3,3,3,3,2,2,2,2,5,5,1,1,1,3,5,5,5,5\
        ,6,6,6,8,5,4,4,4,6,3,3,3,3,3,3,2,2,2,3,1]#小毛驢

music3=[1,2,3,1,1,2,3,1,3,4,5,3,4,5,5,6,5,4,3,1,5,6,5,4,3,1,2,3,1,2,3,1]#兩隻老虎
counter2=0
number=[]
for i in range(21):
    number.append(i+1)
#     print(number)
# for note in Frequency:
#     wd.Beep(note, 500)
pg.init()
clock = pg.time.Clock()      
width, height = 960, 500                      
screen = pg.display.set_mode((width, height))   
pg.display.set_caption("小鋼琴") 
dfont = pg.font.SysFont("simsunnsimsun", 20)    #下方訊息字體
counter=0
bg = pg.Surface(screen.get_size())
bg = bg.convert()
bg.fill((255,255,255))
running = True
playing = False
# one=pg.image.load("0.png")
# two=pg.image.load("4.png")
downmsg = "請點擊滑鼠左鍵以運行程式"
startmsg = "遊戲說明"
piano = []
k=0
num=0
for i in range(22):
        string='i.png'
        string=string.replace("i",str(int(i)))
        piano.append(pg.image.load(string))
ControlPic = []
ControlPic.append(pg.image.load('control.png'))
def ChoosePic(n):
    for i in range(22):
        if i==n:
            screen.blit(piano[i],(0,0))
while running:
    clock.tick(60)        #每秒執行30次
    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False
    message = dfont.render(downmsg, 1, (255,0,255))
    message2 = dfont.render(startmsg, 1, (0,0,0))
    buttons = pg.mouse.get_pressed()  #檢查滑鼠按鈕
    if buttons[0]:            #按滑鼠左鍵後球可移動       
        playing = True
    if playing ==False:
        screen.blit(ControlPic[0],(0,0))
    if playing==True:
        downmsg=""
        keys = pg.key.get_pressed()
        if keys[pg.K_1]:
            num=1
            wd.Beep(Frequency[1], 500)
        if keys[pg.K_2]:
            wd.Beep(Frequency[2], 500)
            num=2
        if keys[pg.K_3]:
            wd.Beep(Frequency[3], 500)
            num=3
        if keys[pg.K_4]:
            wd.Beep(Frequency[4], 500)
            num=4
        if keys[pg.K_5]:
            wd.Beep(Frequency[5], 500)
            num=5
        if keys[pg.K_6]:
            wd.Beep(Frequency[6], 500) 
            num=6
        if keys[pg.K_7]:
            wd.Beep(Frequency[7], 500)
            num=7
        if keys[pg.K_q]:
            num=8
            wd.Beep(Frequency[8], 500)
        if keys[pg.K_w]:
            wd.Beep(Frequency[9], 500)
            num=9
        if keys[pg.K_e]:
            wd.Beep(Frequency[10], 500)
            num=10
        if keys[pg.K_r]:
            wd.Beep(Frequency[11], 500)
            num=11
        if keys[pg.K_t]:
            wd.Beep(Frequency[12], 500)
            num=12
        if keys[pg.K_y]:
            wd.Beep(Frequency[13], 500) 
            num=13
        if keys[pg.K_u]:
            wd.Beep(Frequency[14], 500)
            num=14
        if keys[pg.K_a]:
            num=15
            wd.Beep(Frequency[15], 500)
        if keys[pg.K_s]:
            wd.Beep(Frequency[16], 500)
            num=16
        if keys[pg.K_d]:
            wd.Beep(Frequency[17], 500)
            num=17
        if keys[pg.K_f]:
            wd.Beep(Frequency[18], 500)
            num=18
        if keys[pg.K_g]:
            wd.Beep(Frequency[19], 500)
            num=19
        if keys[pg.K_h]:
            wd.Beep(Frequency[20], 500) 
            num=20
        if keys[pg.K_j]:
            wd.Beep(Frequency[21], 500)
            num=21
        if keys[pg.K_0]:
            for m in music:
                wd.Beep(Frequency[m], 500)
                time.sleep(0.05)
                ChoosePic(m)
                pg.display.update()
        if keys[pg.K_o]:
            for m in music3:
                if(counter2==14):
                    time.sleep(0.3)
                wd.Beep(Frequency[m], 500)
                counter2+=1
                time.sleep(0.05)
                ChoosePic(m)
                pg.display.update()
        if keys[pg.K_p]:
            for m in music2:
                counter+=1
                if(counter==14):
                    time.sleep(0.3)
                if(counter==27):
                    time.sleep(0.3)
                if(counter==41):
                    time.sleep(0.3)
                if(counter>44 and counter<51):
                    wd.Beep(Frequency[m], 300)
                else:
                    wd.Beep(Frequency[m], 500)
                    time.sleep(0.05)
                ChoosePic(m)
                pg.display.update()
        ChoosePic(num)
        time.sleep(0.01) #緩衝圖片
        num=0
    #screen.blit(bg, (0,0))  #重繪視窗
    
   
#     counter=counter+1
#     if(counter==2):
#         num=0
#         counter=0
#     screen.blit(piano[0],(0,0))
#     screen.blit(message, (screen.get_width()/2-125,screen.get_height()-30))
#     screen.blit(message2, (screen.get_width()/2-50,0))
    pg.display.update()
del dfont #此行很重要 要不然執行第二次會掛掉
pg.quit()
```
