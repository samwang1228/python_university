```python
import random
from random import choice
import pygame
from pygame import mixer

class Deck:
    def __init__(self,font):
        #font
        self.font=font
        #Card
        self.A=[]
        for i in range(53):
            self.A.append(i)
        self.A.remove(0)
        self.Card=[]
        for i in range(53):
            string='i.png'
            string=string.replace("i",str(int(i)))
            self.Card.append(pygame.image.load(string))
        self.num=52
        #player
        self.PlayerCard=[]
        self.PlayerCard_posX=[]
        self.PlayerCard_posX.append(20)
        self.PlayerCard_posX.append(130)
        self.PlayerCard_posY=500
        self.PlayerCard_sum=0
        self.PlayerCard_num=0
        self.Player_has_nA=0
        #dealer
        self.DealerCard=[]
        self.DealerCard_posX=[]
        self.DealerCard_posX.append(770)
        self.DealerCard_posX.append(660)
        self.DealerCard_posY=70
        self.hiddenCard=self.Card[0]
        self.DealerCard_sum=0
        self.DealerCard_num=0
        self.Dealer_has_nA=0
    
    def Deal(self):
        for i in range(2):
            p=choice(self.A)
            self.A.remove(p)
            self.PlayerCard.append(self.Card[p])
            self.num-=1
            if (p%13==0 or p%13==11 or p%13==12):
                self.PlayerCard_sum+=10
            elif p%13==1:
                self.PlayerCard_sum+=11
                self.Player_has_nA+=1
            else:
                self.PlayerCard_sum+=p%13
            self.PlayerCard_num+=1
        self.Player_change_A(p)
        d=choice(self.A)
        self.A.remove(d)
        self.DealerCard.append(self.Card[0])
        self.hiddenCard=self.Card[d]
        self.num-=1
        if (d%13==0 or d%13==11 or d%13==12):
                self.DealerCard_sum+=10
        elif d%13==1:
            self.DealerCard_sum+=11
            self.Dealer_has_nA+=1
        else:
            self.DealerCard_sum+=d%13
        self.DealerCard_num+=1
        d=choice(self.A)
        self.A.remove(d)
        self.DealerCard.append(self.Card[d])
        self.num-=1
        if (d%13==0 or d%13==11 or d%13==12):
            self.DealerCard_sum+=10
        elif d%13==1:
            self.DealerCard_sum+=11
            self.Dealer_has_nA+=1
        else:
            self.DealerCard_sum+=d%13
        self.Dealer_change_A(d)
        self.DealerCard_num+=1
        
    def PlayerHit(self):
        p=choice(self.A)
        self.A.remove(p)
        self.PlayerCard.append(self.Card[p])
        self.num-=1
        self.PlayerCard_posX.append(self.PlayerCard_posX[self.PlayerCard_num-1]+110)
        self.PlayerCard_num+=1
        if (p%13==0 or p%13==11 or p%13==12):
            self.PlayerCard_sum+=10
        elif p%13==1:
            self.PlayerCard_sum+=11
            self.Player_has_nA+=1
        else:
            self.PlayerCard_sum+=p%13
        return p
        
    def DealerHit(self):
        d=choice(self.A)
        self.A.remove(d)
        self.DealerCard.append(self.Card[d])
        self.num-=1
        self.DealerCard_posX.append(self.DealerCard_posX[self.DealerCard_num-1]-110)
        self.DealerCard_num+=1
        if (d%13==0 or d%13==11 or d%13==12):
            self.DealerCard_sum+=10
        elif d%13==1:
            self.DealerCard_sum+=11
            self.Dealer_has_nA+=1
        else:
            self.DealerCard_sum+=d%13
        return d
        
    def BlackJack(self):
        if (self.PlayerCard_sum==21) and (self.DealerCard_sum!=21):
            self.DealerCard[0]=self.hiddenCard
            win_text=self.font.render("Player get Blackjack",True,(255,255,255))
            screen.blit(win_text,(10,660))
            return False
        elif (self.PlayerCard_sum!=21) and (self.DealerCard_sum==21):
            self.DealerCard[0]=self.hiddenCard
            lose_text=self.font.render("Dealer get Blackjack",True,(255,255,255))
            screen.blit(lose_text,(10,660))
            return False
        else:
            return True
    
    def Player_change_A(self,p):
        if self.Player_has_nA!=0 and self.PlayerCard_sum>21:
            self.PlayerCard_sum-=10
            self.Player_has_nA-=1
    
    def Dealer_change_A(self,d):
        if self.Dealer_has_nA!=0 and self.DealerCard_sum>21:
            self.DealerCard_sum-=10
            self.Dealer_has_nA-=1
    
    def bust(self):
        if self.PlayerCard_sum>21:
            bust_text=self.font.render("Player busted", True, (255, 255, 255))
            screen.blit(bust_text, (10, 655))
        elif self.DealerCard_sum>21:
            bust_text = self.font.render("Dealer busted", True, (255, 255, 255))
            screen.blit(bust_text, (10, 655))
        else:
            return
            
    def judge(self):
        if (self.PlayerCard_sum==21) and (self.DealerCard_sum!=21) and (self.PlayerCard_num==2):
            self.DealerCard[0]=self.hiddenCard
            win_text=self.font.render("Player get Blackjack",True,(255,255,255))
            screen.blit(win_text,(10,655))
            return 1
        elif (self.PlayerCard_sum!=21) and (self.DealerCard_sum==21) and (self.DealerCard_num==2):
            self.DealerCard[0]=self.hiddenCard
            lose_text=self.font.render("Dealer get Blackjack",True,(255,255,255))
            screen.blit(lose_text,(10,655))
            return 2
        elif self.PlayerCard_sum>21:
            bust_text=self.font.render("Player busted", True, (255, 255, 255))
            screen.blit(bust_text, (10, 655))
            return 3
        elif self.DealerCard_sum>21:
            bust_text = self.font.render("Dealer busted", True, (255, 255, 255))
            screen.blit(bust_text, (10, 655))
            return 4
        elif self.PlayerCard_sum>self.DealerCard_sum:
            win_text=self.font.render("Player win",True,(255,255,255))
            screen.blit(win_text,(10,655))
            return 5
        elif self.PlayerCard_sum<self.DealerCard_sum:
            lose_text=self.font.render("Player lose",True,(255,255,255))
            screen.blit(lose_text,(10,655))
            return 6
        elif self.PlayerCard_sum==self.DealerCard_sum:
            push_text=self.font.render("The game pushed",True,(255,255,255))
            screen.blit(push_text,(10,655))
            return 7

    def end(self):
        fin_text=self.font.render("You lose, and you have no bet. Want to try again? Press Y",True,(255,255,255))
        screen.blit(fin_text,(10,655))

    def show(self):
        for i in range(self.PlayerCard_num):
            screen.blit(self.PlayerCard[i], (self.PlayerCard_posX[i], self.PlayerCard_posY))
        for i in range(self.DealerCard_num):
            screen.blit(self.DealerCard[i], (self.DealerCard_posX[i], self.DealerCard_posY))
        Psum=self.font.render(str(self.PlayerCard_sum), True, (255, 255, 255))
        screen.blit(Psum,(10,450))
        
    def Dealer_card_show(self):
        self.DealerCard[0]=self.hiddenCard
            
        
        
# Intialize the pygame
pygame.init()

# create the screen
screen = pygame.display.set_mode((1200, 700))

# Background
background = pygame.image.load('bjs.png')

# Caption and Icon
pygame.display.set_caption("black jack")
icon = pygame.image.load('spade_ace.jpg')
pygame.display.set_icon(icon)

# Font setting
font = pygame.font.Font('freesansbold.ttf', 40)

# Chips
chips=2000.0
bet=100.0
score_font=pygame.font.Font('freesansbold.ttf', 32)

def Show_chip_and_bet(chips,bet):
    chips=score_font.render("Chips : " + str(chips), True, (255, 255, 255))
    screen.blit(chips, (930, 590))
    bets=score_font.render("Bet : " + str(bet),True,(255,255,255))
    screen.blit(bets,(930,620))

# Rule
rule_font=pygame.font.Font('freesansbold.ttf', 24)
def show_rule():
    Deal=rule_font.render("A : Deal",True,(255,255,255))
    Hit=rule_font.render("SPACE : Hit",True,(255,255,255))
    Stand=rule_font.render("Z : Stand",True,(255,255,255))
    Double=rule_font.render("X : Double",True,(255,255,255))
    Clear=rule_font.render("L_CTRL : Clear",True,(255,255,255))
    higher=rule_font.render("UP : +bet",True,(255,255,255))
    lower=rule_font.render("DOWN : -bet",True,(255,255,255))
    all_in=rule_font.render("S : All_in",True,(255,255,255))
    screen.blit(Deal,(995,300))
    screen.blit(Hit,(995,330))
    screen.blit(Stand,(995,360))
    screen.blit(Double,(995,390))
    screen.blit(Clear,(995,420))
    screen.blit(higher,(995,450))
    screen.blit(lower,(995,480))
    screen.blit(all_in,(995,510))
    
running=True
while running:
    D=Deck(font)
    screen.fill((0, 0, 0))
    screen.blit(background, (0, 0))
    Show_chip_and_bet(chips,bet)
    show_rule()
    pygame.display.update()
    Player_turn=False
    Dealer_turn=False
    judge_turn=False
    score_cal=False
    Double=False
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                if bet+100<=chips:
                    bet+=100
            elif event.key == pygame.K_DOWN:
                if bet-100>=100:
                    bet-=100
            elif event.key == pygame.K_s:
                bet=chips
            elif event.key == pygame.K_a:
                screen.fill((0, 0, 0))
                screen.blit(background, (0, 0))
                D.Deal()
                D.show()
                det=D.BlackJack()
                Show_chip_and_bet(chips,bet)
                show_rule()
                pygame.display.update()
                Player_turn=det
                Dealer_turn=det
                judge_turn=True
                score_cal=True
        
    while Player_turn and D.PlayerCard_sum<21:
        for event in pygame.event.get():
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_x:
                    if D.PlayerCard_sum==11 and D.PlayerCard_num==2:
                        screen.fill((0, 0, 0))
                        screen.blit(background, (0, 0))
                        p=D.PlayerHit()
                        D.Player_change_A(p)
                        D.show()
                        D.bust()
                        bet*=2
                        Show_chip_and_bet(chips,bet)
                        show_rule()
                        pygame.display.update()
                        Double=True
                        Player_turn=False
                if event.key == pygame.K_SPACE:
                    screen.fill((0, 0, 0))
                    screen.blit(background, (0, 0))
                    p=D.PlayerHit()
                    D.Player_change_A(p)
                    D.show()
                    D.bust()
                    Show_chip_and_bet(chips,bet)
                    show_rule()
                    pygame.display.update()
                if event.key == pygame.K_z:
                    Player_turn=False

    while Dealer_turn:
        D.Dealer_card_show()
        if D.PlayerCard_sum>21:
            Dealer_turn=False
        elif D.DealerCard_sum<17:
            screen.fill((0, 0, 0))
            screen.blit(background, (0, 0))
            d=D.DealerHit()
            D.Dealer_change_A(d)
            D.show()
            D.bust()
            Show_chip_and_bet(chips,bet)
            show_rule()
            pygame.display.update()
        else:
            Dealer_turn=False

    while judge_turn:
        screen.fill((0, 0, 0))
        screen.blit(background, (0, 0))
        s=D.judge()
        D.show()
        Show_chip_and_bet(chips,bet)
        show_rule()
        pygame.display.update()
        if score_cal:
            if s==1:
                chips+=bet*1.5
            elif s==2:
                chips-=bet*1.5
            elif s==3:
                chips-=bet
            elif s==4:
                chips+=bet
            elif s==5:
                chips+=bet
            elif s==6:
                chips-=bet
            score_cal=False
        if Double:
            bet/=2
            Double=False
        if chips<=0:
            judge_turn=False
        for event in pygame.event.get():
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LCTRL:
                    judge_turn=False
    
    if bet>chips and chips>=0:
        bet=chips
    elif chips<0 and bet>chips:
        bet=0
    while chips<=0 and running:
        screen.fill((0, 0, 0))
        screen.blit(background, (0, 0))
        D.end()
        D.show()
        Show_chip_and_bet(chips,bet)
        show_rule()
        pygame.display.update()
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_y:
                    chips=2000
                    bet=100
    pygame.display.update()
pygame.quit()
```
