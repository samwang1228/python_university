#  鐘型函數

```python
import numpy as np
import matplotlib.pyplot as pl 
x=np.linspace(-3,3,1001)
fx=np.exp(-x**2)
pl.plot(x,fx)
pl.grid()
```

```python
import sympy as sy 
import sympy.plotting as pl 
x=sy.symbols('x')
fx=sy.exp(-x**2)
pl.plot(fx,(x,-3,3))
```

```python
import numpy as np
import matplotlib.pyplot as pl 
π = np.pi 
x = np.linspace(-10,10,10001)

fx=np.exp(-x**2/2)/(2*π)**0.5
pl.plot(x,fx)
pl.grid()
```
