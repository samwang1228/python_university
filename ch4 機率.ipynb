# 自由度

```python
x=np.array([1.9,2.4,3.0,3.5,4.2])
n=x.size
xBar=x.mean()
s2=x.var(ddof=1)
n , xBar , s2
```

```python
σ=1
prob=st.chi2.cdf(x=s2,df=n-1,scale=σ**2/(n-1))
prob
```
