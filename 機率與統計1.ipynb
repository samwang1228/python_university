# 統計

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as pl
import pandas as pd

X0=[ .32, .53, .28, .37, .47,
     .43, .36, .42, .38, .43]
X1=[ .26, .43, .47, .49, .52,
     .75, .79, .86, .62, .46]

X0= np.array(X0)
X1= np.array(X1)

pdX= pd.DataFrame(
        {'X0, noNitrigen': X0,
         'X1, withNitrogen':X1}
        )
```

```python
# pl.tick_params(axis='x',colors='red')
# pl.tick_params(axis='y',colors='red')
pdX.plot.line(grid=True, marker='o')
pdX.plot.bar(grid=True)

pdX.plot.barh(grid=True)

pdX.plot.hist(grid=True)
pdX.plot.hist(grid=True, bins=100)

pdX.plot.density(grid=True)

pdX.plot.scatter(x=0, y=1, grid=True)
```

```python
describeX= pdX.describe()
describeX
```

```python
pdX.boxplot(grid=True)
```

```python

# standard error from mean (sem)

st.sem(X0), st.sem(X1)
```

```python
# 95% Confidence Inteval (信心區間)

mu0Interval= st.t.interval(0.95, 
                           len(X0)-1, 
                           loc=np.mean(X0), 
                           scale=st.sem(X0))
mu1Interval= st.t.interval(0.95, 
                           len(X1)-1, 
                           loc=np.mean(X1), 
                           scale=st.sem(X1))
mu0Interval, mu1Interval
```

```python

# hypothesis testing

st.ttest_ind(X0,X1)
```
