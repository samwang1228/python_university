```python
import pandas_datareader.data as web 
```

```python
df=web.DataReader("AAPL","yahoo",start="2012-9-12",end="2020-7-20")
df.head()
```

```python
p=df["Adj Close"]
p.head()
```

```python
p.plot()
```

```python
from matplotlib import pyplot as plt 
plt.title("Apple-stock ROI") 
r=p.diff()/p #投資報酬率
r[-100:].plot() #只看前100天
```
